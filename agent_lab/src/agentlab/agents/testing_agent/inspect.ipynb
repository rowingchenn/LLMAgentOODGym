```python
import os
import gzip
import pickle
import json
# 打印step
file_path = '/home/weichen/AgentLab_OOD/src/agentlab/agents/testing_agent/test_1/2024-10-30_23-13-08_GenericAgent-gpt-4-1106-preview_on_workarena.servicenow.create-incident_484/step_5.pkl.gz'
if os.path.getsize(file_path) == 0:
    print("文件为空")
else:
    with gzip.open(file_path, 'rb') as file:
        data = pickle.load(file)
        print(vars(data))
        # 将数据保存到当前目录下的 JSON 文件中
        output_file_path = './step_5_uncompressed.json'
        
        # 自定义序列化函数
        def custom_serializer(obj):
            if hasattr(obj, '__dict__'):
                return obj.__dict__
            else:
                return str(obj)
        
        with open(output_file_path, 'w') as output_file:
            json.dump(data, output_file, default=custom_serializer, indent=4)
        
        print(f"数据已保存到 {output_file_path}")
```
